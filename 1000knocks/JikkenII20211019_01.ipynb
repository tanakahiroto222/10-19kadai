# 04. 元素記号
“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```python
a = "Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
a = a.replace(".", "")
a = a.replace(",", "")
b = a.split()
dic = {}
for i, word in enumerate(b):
    if i+1 == 1 or 5 or 6 or 7 or 8 or 9 or 15 or 16 or 19:
        dic[word[:2]] = i+1
    else:
        dic[word[:3]] = i+1
print(dic)
```

# 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．


n-gram とは n 個のデータの連接である．  
1-gram, 2-gram, 3-gram, ..., n-gram となる．
一般に 1-gram, 2-gram, 3-gram をそれぞれ uni-gram, bi-gram, tri-gram と書く．  
それぞれ「ユニグラム」，「バイグラム」，「トライグラム」と読む．  
単語 n-gram は単語を単位とした連接，文字 n-gram は文字を単位とした連接である．  
"I saw a girl with a telescope." という文があったとしたら，  
単語 uni-gram は "I", "saw", "a",...,"telescope."，  
文字 uni-gram は "I", " ", "s", "a", "w",...,"."，  
単語 bi-gram は "I saw", "saw a", "a girl",..., "a telescope."，  
文字 bi-gram は "I ", " s", "sa", "aw", "w ", " a",...,"e."  
となる．

```python
def make_ngram(sequence, n=1, w=True): #w==Trueなら単語n-gram, Falseなら文字n-gram
    if w == True:
        l = sequence.split()
    else:
        l = list(sequence)
    ngram = []
    for i in range(len(l) - (n-1)):
        x = l[i]
        for j in range(1, n):
            if w == True:
                x += " "
            x += l[i+j]
        ngram.append(x)
    return ngram

a = "I am an NLPer"
print(a)
print("単語bi-gramは", make_ngram(a, 2, True))
print("文字bi-gramは", make_ngram(a, 2, False))
```

# 06. 集合
“paraparaparadise”と”paragraph”に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，’se’というbi-gramがXおよびYに含まれるかどうかを調べよ．


```python
x = "paraparaparadise"
y = "paragraph"
X = set(make_ngram(x, 2, False))
Y = set(make_ngram(y, 2, False))
print("X", X)
print("Y", Y)
print("XとYの和集合", X | Y)
print("XとYの積集合", X & Y)
print("XとYの差集合", X - Y)
print("'se'がXに含まれるか", "se" in X)
print("'se'がYに含まれるか", "se" in Y)
```

# 08. 暗号文
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

英小文字ならば(219 - 文字コード)の文字に置換
その他の文字はそのまま出力
この関数を用い，英語のメッセージを暗号化・復号化せよ．

```python
def cipher(string):
    s = list(string)
    new_s = []
    for c in s:
        if c.islower():
            new_s.append(chr(219 - ord(c)))
        else:
            new_s.append(c)
    return "".join(new_s)

str = "Hello I'm hungry."
print("元の文字列", str)
ciphered = cipher(str)
print("暗号化", ciphered)
print("復号化", cipher(ciphered))
```

# 09. Typoglycemia
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ．

```python
import random
s = "I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
l = s.split()
new_l = []
for word in l:
    if len(word) <= 4:
        new_l.append(word)
    else:
        new_l.append(word[0] + "".join(random.sample(word[1:-1], len(word)-2)) + word[-1])
new_s = " ".join(new_l)
print(s)
print(new_s)
```
